```python
def AGE_BUCKET(Age):
    if Age < 30:
        return 'Less than 30'
    elif Age >= 30 and Age <= 50:
        return '30 to 50'
    elif Age >= 51 and Age <= 70:
        return '51 to 70'
    else:
        return '70+'
    
```

```python
MED_SELECT['AGE_BUCKET']=MED_SELECT['Age'].apply(AGE_BUCKET)
```

```python
MED_SELECT
```

```python
def SPENT_BUCKET (Spent_amount):
    if Spent_amount < 300:
        return 'Less than 300'
    elif Spent_amount >= 301 and Spent_amount <= 500:
        return '301 to 500'
    elif Spent_amount >= 501 and Spent_amount <= 700:
        return '501 to 700'
    elif Spent_amount >= 701 and Spent_amount <= 1000:
        return '701 to 1000'
    else:
        return '1000+'
```

```python
MED_SELECT['SPENT_BUCKET']=MED_SELECT['Spent amount'].apply(SPENT_BUCKET)
```

```python
def CUSTOMER_SEGMENT(row):
    Age = row['Age']
    GENDER = row['GENDER']
    Spent_amount = row['Spent amount']
    
    if 30 <= Age <= 70 and Spent_amount >= 300:
        if GENDER == 'Female':
            return 'PREMIER'
        elif GENDER == 'Male':
            return 'PREMIER'
    return 'NON-PREMIER'

MED_SELECT_V6['CUSTOMER_SEGMENT'] = MED_SELECT_V6[['Age', 'GENDER', 'Spent amount']].apply(CUSTOMER_SEGMENT, axis=1)
```

```python
def DISCOUNT(row):
    COURSE = row['COURSE']
    TOOL = row['TOOL']
    Fee = row['FEE']
    
    if COURSE == 'MPIDS':
        if TOOL == 'SAS' and Fee == 25000:
            return 20
        elif TOOL == 'POWER BI' and Fee == 19000:
            return 30
        elif TOOL == 'ADVANCED ANALYTICS' and Fee == 45000:
            return 10
        elif TOOL == 'MACHINE LEARNING' and Fee == 70000:
            return 20
    elif COURSE == 'ADA':
        if TOOL == 'SQL' and Fee == 15000:
            return 10
        elif TOOL == 'PYTHON' and Fee == 15000:
            return 10
        elif TOOL == 'SAS' and Fee == 25000:
            return 20
        elif TOOL == 'R' and Fee == 15000:
            return 10
        elif TOOL == 'ADVANCED ANALYTICS' and Fee == 45000:
            return 10
    elif COURSE == 'DSA' and TOOL == 'SAS' and Fee == 25000:
        return 10
    elif COURSE == 'MINI MODULE':
        if TOOL == 'TABLEAU' and Fee == 17000:
            return 10
        elif TOOL == 'POWER BI' and Fee == 19000:
            return 10
        
    return 0

STU_MASTER['DISCOUNT'] = STU_MASTER.apply(DISCOUNT, axis=1)
```
