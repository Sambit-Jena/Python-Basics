```python
import pandas as pd
task_8_data=pd.read_excel(r'C:\Users\SAMBIT KUMAR\Downloads\A5.ASSIGNMENTS-20240408T055501Z-001\A5.ASSIGNMENTS\PYTHON SESSION-3 ASSIGNMENTS.xlsx', sheet_name='Sheet13')
task_8_data
```

```python
STEP_1=task_8_data.groupby(['COUNTRY','PRODUCT'], as_index=False).agg(TOTAL_UNITS_SOLD=('UNITS','sum'))
STEP_1
```

```python
STEP_2=STEP_1.groupby(['COUNTRY'],as_index=False).agg(AVG_UNIT_SOLD=('TOTAL_UNITS_SOLD','mean'))
STEP_2
```

```python
STEP_3=pd.merge(left=STEP_1,right=STEP_2, how='left', on='COUNTRY')
STEP_3
```

```python
def CATEGORY(row):
    total_spent= row['TOTAL_UNITS_SOLD']
    average_spent = row['AVG_UNIT_SOLD']
    if total_spent >= 1.25*average_spent:
        return 'CAT-A'
    elif total_spent >= 0.5*average_spent:
        return 'CAT-B'
    else:
        return 'CAT-C'
        

STEP_3['CATEGORY'] = STEP_3[['TOTAL_UNITS_SOLD', 'AVG_UNIT_SOLD']].apply(CATEGORY, axis=1)
```

```python
STEP_4=STEP_3.groupby(['COUNTRY','CATEGORY'], as_index=False).agg({'TOTAL_UNITS_SOLD':'sum'})
STEP_4
```

```python
STEP_5=STEP_4.pivot(index='COUNTRY',columns='CATEGORY',values='TOTAL_UNITS_SOLD')
STEP_5.fillna('')
```

```python
STEP_5['CAT-A_in_%']=(STEP_5['CAT-A']/STEP_5.sum(axis=1))*100
STEP_5['CAT-B_in_%']=(STEP_5['CAT-B']/STEP_5.sum(axis=1))*100
STEP_5['CAT-C_in_%']=(STEP_5['CAT-C']/STEP_5.sum(axis=1))*100
STEP_5
```

```python
STEP_6=STEP_5[['CAT-A_in_%','CAT-B_in_%','CAT-C_in_%']]

STEP_6['CAT-A_in_%'] = STEP_6['CAT-A_in_%'].apply(lambda x: f'{x:,.0f}%')
STEP_6['CAT-B_in_%'] = STEP_6['CAT-B_in_%'].apply(lambda x: f'{x:,.0f}%')
STEP_6['CAT-C_in_%'] = STEP_6['CAT-C_in_%'].apply(lambda x: f'{x:,.0f}%')
STEP_6.replace('nan%', '0%', inplace=True)

STEP_6
```
