```python
FINAL_DATA['AVG NO OF VISITS']=FINAL_DATA['NO_OF_VISIT'].mean()
FINAL_DATA['AVG SPENT']=FINAL_DATA['SPENT'].mean()
FINAL_DATA['AVG SPENT_PER_VISIT']=FINAL_DATA['SPENT PER VISIT'].mean()
FINAL_DATA
```

```python
def VISITS_SEGMENTATION(y):
    visits=y["NO_OF_VISIT"]
    avg_visit=y["AVG NO OF VISITS"]
    
    if visits>=avg_visit:
        return "PREMIER"
    else:
        return "NON-PREMIER"
    
```

```python
FINAL_DATA['CUSTOMER VISITS_SEGMENTATION']=FINAL_DATA.apply(VISITS_SEGMENTATION, axis=1)
FINAL_DATA
```

```python
def SPENT_SEGMENTATION(y):
    spents=y["SPENT"]
    avg_spents=y["AVG SPENT"]
    
    if spents>=avg_spents:
        return "PREMIER"
    else:
        return "NON-PREMIER"
```

```python
FINAL_DATA['CUSTOMER SPENT_SEGMENTATION']=FINAL_DATA.apply(SPENT_SEGMENTATION, axis=1)
FINAL_DATA
```

```python
def SPENT_PER_VISIT_SEGMENTATION(y):
    spents_per_visit=y["SPENT PER VISIT"]
    avg_spents_per_visit=y["AVG SPENT_PER_VISIT"]
    
    if spents_per_visit>=avg_spents_per_visit:
        return "PREMIER"
    else:
        return "NON-PREMIER"
```

```python
FINAL_DATA['CUSTOMER SPENT_PER_VISIT_SEGMENTATION']=FINAL_DATA.apply(SPENT_PER_VISIT_SEGMENTATION, axis=1)
FINAL_DATA
```

```python
def CUSTOMER_SEGMENT(y):
    visits = y["CUSTOMER VISITS_SEGMENTATION"]
    spents=y["CUSTOMER SPENT_SEGMENTATION"]
    spents_per_visit=y["CUSTOMER SPENT_PER_VISIT_SEGMENTATION"]
    
    if visits=="PREMIER" and spents=="PREMIER" and spents_per_visit=="PREMIER":
        return "BONAFIED"
    elif visits=="PREMIER" and spents=="PREMIER" and spents_per_visit=="NON-PREMIER":
        return "POWER"
    elif visits=="NON-PREMIER" and spents=="PREMIER" and spents_per_visit=="PREMIER":
        return "CASUAL"
    elif visits=="NON-PREMIER" and spents=="NON-PREMIER" and spents_per_visit=="NON-PREMIER":
        return "MISER"
    elif visits=="PREMIER" and spents=="PREMIER" and spents_per_visit=="PREMIER":
        return "POWER"
    elif visits=="NON-PREMIER" and spents=="PREMIER" and spents_per_visit=="PREMIER":
        return "CASUAL"
    else:
        return "POWER"
```

```python
FINAL_DATA['CUSTOMER SEGMENT']=FINAL_DATA.apply(CUSTOMER_SEGMENT, axis=1)
FINAL_DATA
```
